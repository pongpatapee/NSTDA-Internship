```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
col_vals = 's1 s2 s3 s4 s5 s6 s7 s8'.split()
```

```python
control = pd.read_csv('data/control.csv', names=col_vals)
alcohol = pd.read_csv('data/alcohol.csv', names=col_vals)
lotion = pd.read_csv('data/body_lotion.csv', names=col_vals)

control['target'] = 'control'
alcohol['target'] = 'alcohol'
lotion['target'] = 'lotion'
```

```python
df = pd.concat([control, alcohol, lotion])
```

```python
sns.pairplot(df, hue='target')
```

```python
df.to_csv('training_df.csv', index=False, header=False)
```
